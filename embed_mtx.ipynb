```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv("pca.csv")
df
```

```python
pca_vec = df["pca_vec"].values.tolist()
pca_vec[0]
```

```python
usr_vct = df["clustr_id"].values.tolist()
aid_vct = df["aid"].values.tolist()
```

```python
from ast import literal_eval

U = np.zeros((100,10))
V = np.zeros((91,10))

cnt_u = [0] * 100
cnt_v = [0] * 91

for i in range(0,len(df)):
    cnt_u[usr_vct[i]] += 1
    cnt_v[aid_vct[i]] += 1
    embed = literal_eval(pca_vec[i])
    for j in range(0,10):
        U[usr_vct[i]][j] += embed[j]
        V[aid_vct[i]][j] += embed[j]

for i in range(0,100):
    for j in range(0,10):
        U[i][j] /= cnt_u[i]

for i in range(0,91):
    for j in range(0,10):
        V[i][j] /= cnt_v[i]

U
```

```python
with open('U_bert.plk', '+bw') as f:
    pickle.dump(U, f)
```

```python
with open('V_bert.plk', '+bw') as f:
    pickle.dump(V, f)
```
