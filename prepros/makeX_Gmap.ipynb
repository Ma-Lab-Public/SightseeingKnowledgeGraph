```python
import numpy as np
import pickle
import pandas as pd
```

```python
cluster = pd.read_csv("classify_id.csv", index_col=0)
cluster
```

```python
trj = pd.read_csv("classify_latest.csv")
trj = trj.sort_values(['clustr_id', 'Date'])
trj
```

```python
X = np.zeros((100,91))

uid_list = cluster.index.tolist()
cnt_list = [0] * 91

for aid in trj['aid'].values.tolist():
    cnt_list[aid] += 1

# cnt_list
aoi_list = cluster['aid'].values.tolist()
# aoi_list
for i in range(0, len(uid_list)):
    trj_list = aoi_list[i].split(',')
    for j in range(0,len(trj_list)):
        X[i][int(trj_list[j])] += 1

for i in range(0,100):
    for j in range(0,91):
        X[i][j] /= cnt_list[j]
        X[i][j] *=100
X
```

```python
# check 0
for i in range(0,100):
    sum = 0
    for j in range(0,91):
        sum += X[i][j]
    if sum == 0:
        print("empty line!")
```

```python
with open('X_Gmap.plk', '+bw') as f:
    pickle.dump(X, f)
```
