```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# 读取数据
df = pd.read_csv('data/all.csv')

# 删除无用列
# cols_to_drop = [1, 2, 3, 4, 5, 6, 7]
# df.drop(df.columns[cols_to_drop], axis=1, inplace=True)

# 提取 tensor 列，转为二维 numpy 数组
vector_list = df['tensor'].tolist()
vec_2d_array = []


for item in vector_list:
    # 安全清洗字符串
    cleaned = item.replace('tensor(', '').replace(')', '').replace('[', '').replace(']', '')
    array_data = np.fromstring(cleaned, sep=',')
    vec_2d_array.append(array_data)

vec_2d_array = np.array(vec_2d_array)  # shape: (9004, embedding_dim)

# KMeans 聚类
num_clusters = 50
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(vec_2d_array)

# 保存聚类结果
df['cluster_id'] = cluster_labels

df['vec_2d'] = vec_2d_array.tolist()

# 输出前几行验证
print(df[['cluster_id']].head())
df
```

```python
df.to_csv('clustered_50.csv', index=False)
print("聚类结果已保存到 'clustered_50.csv'")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('clustered_50.csv')

# 删除无用列
df.drop(df.columns[[0,1,2,3,4,5,6,7,8]], axis=1, inplace=True)

# 将 vec_2d 从字符串恢复为数组
def parse_vector(s):
    s = s.strip("tensor()[]")
    return np.fromstring(s, sep=', ')

df['vec_2d'] = df['vec_2d'].apply(parse_vector)

# 计算每个 aid 的平均向量
vec_mean_aid = []
for i in range(91):  # 假设 aid 从 0 到 90
    tmp_df = df[df['aid'] == i]
    if not tmp_df.empty:
        average_vector = np.mean(np.stack(tmp_df['vec_2d'].to_list()), axis=0)
    else:
        average_vector = np.zeros_like(df['vec_2d'].iloc[0])  # 用0向量填充空值
    vec_mean_aid.append(average_vector)

# 构造 DataFrame
df_aid = pd.DataFrame({
    'aid': range(91),
    'vec_mean_aid': vec_mean_aid
})
df_aid
```

```python
import pandas as pd
import numpy as np

# 读取原始聚类数据
df = pd.read_csv('clustered_150.csv')

# 解析 vec_2d 向量列（如果是字符串）
def parse_vector(s):
    s = s.strip("tensor()[]")
    return np.fromstring(s, sep=', ')

df['vec_2d'] = df['vec_2d'].apply(parse_vector)

# 获取所有唯一的 cluster_id
cluster_ids = sorted(df['cluster_id'].unique())

# 计算每个 cluster_id 的平均向量
vec_mean_user = []
for i in cluster_ids:
    tmp_df = df[df['cluster_id'] == i]
    if not tmp_df.empty:
        avg_vec = np.mean(np.stack(tmp_df['vec_2d'].tolist()), axis=0)
    else:
        avg_vec = np.zeros_like(df['vec_2d'].iloc[0])  # 空集则填 0 向量
    vec_mean_user.append(avg_vec)

# 创建一个 DataFrame 保存 cluster_id -> 平均向量 映射
df_cluster_mean = pd.DataFrame({
    'cluster_id': cluster_ids,
    'vec_mean_user': vec_mean_user
})

# 将原始 DataFrame 按 cluster_id 合并平均向量
df_merged = pd.merge(df, df_cluster_mean, on='cluster_id', how='left')

# 保存最终结果
df_merged.to_csv('clustered_150_with_mean.csv', index=False)
print("已将聚类平均向量添加到 clustered_50.csv 并保存为 clustered_150_with_mean.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# 假设 df 是你的原始 DataFrame
df = pd.read_csv('clustered_150_with_mean.csv')
df_cluster = df[['cluster_id', 'vec_mean_user']]
df = df_cluster.sort_values(by='cluster_id', ascending=True)  # 升序排列

# 1. 字符串转向量
def parse_vector(vec_str):
    try:
        vec = np.array([float(x) for x in vec_str.strip("[]").split()])
        return vec if vec.size > 0 else np.nan
    except:
        return np.nan

df['vec_parsed'] = df['vec_mean_user'].apply(parse_vector)

# 2. 过滤无效向量
df_valid = df[df['vec_parsed'].notnull()].copy()

# 3. 构建矩阵
vec_matrix = np.vstack(df_valid['vec_parsed'].to_numpy())

# 4. PCA 降维到 64维
n_components = min(10, vec_matrix.shape[1], vec_matrix.shape[0])  # 防止维度超过样本或原始维度
pca = PCA(n_components=n_components)
vec_reduced = pca.fit_transform(vec_matrix)

# 5. 存入 DataFrame
df_valid['vec_mean_user_64'] = list(vec_reduced)

# 6. 选择 cluster_id + 向量，按 cluster_id 去重
df_cluster_mean = df_valid[['cluster_id', 'vec_mean_user_64']].drop_duplicates(subset=['cluster_id'])

# 输出结果
df_cluster_mean
```

```python
import pandas as pd

# 读取原始数据
df = pd.read_csv('clustered_150_with_mean.csv')
# 合并两个 DataFrame，保留 df_full 的所有行
df_merged = df.merge(df_cluster_mean, on='cluster_id', how='left')

print(f"合并后的数据形状: {df_merged.shape}")

df = df_merged.sort_values(by='cluster_id', ascending=True)  # 升序排列
df.to_csv('clustered_150_PCA.csv', index=False)
print("已将降维平均向量添加到 clustered_150_PCA.csv' 并保存。 ")
# df
```

```python
import pandas as pd

# 读取原始数据
df = pd.read_csv('clustered_150_PCA.csv')

# 保留除 cluster_id 和 vec_mean_user 外的所有列名
cols_to_group = [col for col in df.columns if col not in ['cluster_id', 'vec_mean_user', 'vec_mean_user_64']]

# 分组聚合为列表（保留空值）
grouped_df = df.groupby('cluster_id', dropna=False)[cols_to_group].agg(lambda x: list(x)).reset_index()

# 提取 cluster_id 和唯一对应的 vec_mean_user
vec_mean_df = df[['cluster_id', 'vec_mean_user', 'vec_mean_user_64']].drop_duplicates(subset='cluster_id')

# 合并数据
final_df = pd.merge(vec_mean_df, grouped_df, on='cluster_id', how='left')

# 确保列顺序
cols = ['cluster_id', 'vec_mean_user', 'vec_mean_user_64'] + [col for col in final_df.columns if col not in ['cluster_id', 'vec_mean_user', 'vec_mean_user_64']]
final_df = final_df[cols]

# 排序 cluster_id 升序
final_df = final_df.sort_values(by='cluster_id').reset_index(drop=True)

# 可保存或打印
final_df.to_csv('clustered_150_grouped.csv', index=False)
# print(final_df.head())
final_df
```
